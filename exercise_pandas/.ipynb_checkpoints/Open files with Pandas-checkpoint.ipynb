```python
"""Headers as None are specified as numbers."""
df8 = pandas.read_csv("files/supermarkets-commas.txt", header=None)
df8
```

```python
df8.columns = ['ID','Address','City', 'ZIP','Country','Name','Employees']
df8
```

**Set Index columns**

```python
df8.set_index('ID')
```

```python
df8
```

```python
"""set_index didn't modify permanently the table, so if you want to have a new table you need to assign a new variable"""
df9 = df8.set_index('ID')
df9
```

```python
"""Also you can use the parameter inplace=True"""
df8.set_index('ID', inplace=True)
df8
```

```python
"""When you assign a new index the old index disappear"""
df8.set_index('Address', inplace=True)
df8
```

```python
"""When you use the parameter drop, the index column is also like a data column."""
df8.set_index('Name', inplace=True, drop=False)
df8
```

**Data Analysis Example with geopy**

```python
from geopy.geocoders import ArcGIS
nom = ArcGIS()
```

```python
n = nom.geocode("3995 23rd St, San Francisco, CA 94114")
```

```python
type(n)
```

```python
df = pandas.read_csv("files/supermarkets.csv")
df
```

```python
df["Address"] = df["Address"] + ", " + df["City"] + ", " + df["State"] + ", " + df["Country"]
```

```python
df
```

```python
"""The apply method allows us to use the column data as a parameter for a function. ."""
df["Coordinates"] = df["Address"].apply(nom.geocode)
```

```python
df
```

```python
df.Coordinates
```

```python
df.Coordinates[0]
```

```python
df.Coordinates[0].latitude
```

```python
df.Coordinates[0].longitude
```

```python
df["Latitude"] = df["Coordinates"].apply(lambda x: x.latitude)
df["Longitude"] = df["Coordinates"].apply(lambda x: x.longitude)
df
```
